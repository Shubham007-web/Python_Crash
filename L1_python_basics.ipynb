## 2. Identifiers <a class="anchor" id="id"></a>

- Think of identifiers in Python as **`nametags`** for different things like **classes, functions, or variables**. 
- They're like **containers that hold values**, making it easy to tell one thing apart from another. So, just like you put labels on your toy boxes, Python uses identifiers to keep things organized! 😊


````python
name = "Feynman"
x= 100
data = [24,27,12]
x_1 = 453
````

- `name`, `x`, `data`, and `x_1` are like **nametags (identifiers)** in Python.

These identifiers help us keep track of what's inside each container, just like labels on jars help us know what's inside them.

## Set <a class = 'anchor' id = 'st'></a>
---
-  Set is an **unordered** collection of items, in which there is **no duplicates** are allowed.
-  Indexing and slicing are not allowed in the set because the set is unordered.

````python
s = {1,2,3,-3,5,1,2,2,2}
print(s)
````

````python
# creating empty set
st  = set()
print(st)
````

### Methods and functions of the Set
----
There are some methods and functions of the set:
- `union()`   take the union of the two sets
- `intersection()`  take  the intersection  of the  two sets
- `add()`   add an item in the  set
- `update()` add multiple items in  the set
- `remove()`  removing a element
- `clear()`   deleting all the  items

````python
s1 = {1,2,3,4,5,5,5,6}
s2 = {5,4,10,20,30,40}
print(s1)
print(s2)
````

````python
# union()
s1.union(s2)
````

````python
# intersection
s1.intersection(s2)
````

````python
# add()
s1.add(100)
s1
````

````python
# clear() --> this function wipe all the data inside the set
s1.clear()
print(s1)   # empty set
````

# Operators and Type of operators <a class="anchor" id="op"></a>

In Python, we have tools called **operators** to work with **data (operand)**. Let's check out the types of operators Python has:

---



#### 1. Arithmetic Operators
| Operator | Description    | Example |
|:--------:|:--------------:|:-------:|
| `+`      | Addition       | `x + y` |
| `-`      | Subtraction    | `x - y` |
| `*`      | Multiplication | `x * y` |
| `/`      | Division       | `x / y` |
| `%`      | Modulus        | `x % y` |
| `**`     | Exponentiation | `x ** y`|
| `//`     | Floor Division | `x // y`|

#### 2. Assignment Operators
| Operator | Description                                  | Example |
|:--------:|:--------------------------------------------:|:-------:|
| `=`      | Assigns value from right to left             | `x = y` |
| `+=`     | Adds right operand to left operand           | `x += y`|
| `-=`     | Subtracts right operand from left operand    | `x -= y`|
| `*=`     | Multiplies right operand with left operand   | `x *= y`|
| `/=`     | Divides left operand by right operand        | `x /= y`|
| `%=`     | Takes modulus using left and right operands  | `x %= y`|
| `**=`    | Performs exponential calculation on operands | `x **= y`|
| `//=`    | Performs floor division on operands          | `x //= y`|

#### 3. Relational Operators
| Operator | Description                               | Example |
|:--------:|:-----------------------------------------:|:-------:|
| `==`     | If the values of two operands are equal   | `x == y`|
| `!=`     | If values of two operands are not equal   | `x != y`|
| `>`      | If the value of left operand is greater   | `x > y` |
| `<`      | If the value of left operand is less      | `x < y` |
| `>=`     | If the value of left operand is greater or equal | `x >= y`|
| `<=`     | If the value of left operand is less or equal | `x <= y`|

#### 4. Logical Operators
| Operator | Description   | Example |
|:--------:|:-------------:|:-------:|
| `and`    | Logical AND   | `x and y`|
| `or`     | Logical OR    | `x or y` |
| `not`    | Logical NOT   | `not x`  |

#### 5. Membership Operators
| Operator | Description                                      | Example   |
|:--------:|:------------------------------------------------:|:---------:|
| `in`     | Evaluates to true if it finds a variable         | `x in y`  |
| `not in` | Evaluates to true if it does not find a variable | `x not in y`|
