## `if` statement <a class="anchor" id="if"></a>
---

- **`if` statement**: It's like a green light. If a condition is met, your code takes a specific path.  


## `if...else` statements <a class="anchor" id="ifelse"></a>

- **`if...else` statements**: These are like forks in the road. Depending on a condition, your code chooses one of two paths. 






```python

# Using an if...else statement to check if the number is even or odd

    print("The number is even!")
else:
    print("The number is odd!")

```
---

---
## `if-elif-else` ladder <a class="anchor" id="ladder"></a>

- **`if-elif-else` ladder**: Think of it as a series of checkpoints on a race track. Your code checks conditions one by one and follows the first true one.  

```python

# Using an if-elif ladder to classify the number
if user_number > 10:
    print("The number is greater than 10.")
elif user_number < 10:
    print("The number is less than 10.")
else:
    print("The number is exactly 10.")
    
```

### Q1. Write a program to accept percentages from the user and display the grade according to the following criteria:

| Marks       | Grade |
|-------------|-------|
| Marks > 90  | A+    |
| 80 < Marks ≤ 90 | A    |
| 70 ≤ Marks ≤ 80 | A-   |
| 60 < Marks ≤ 70 | B+   |
| 50 < Marks ≤ 60 | B    |
| 40 < Marks ≤ 50 | B-   |
| 30 < Marks ≤ 40 | C+   |

### Instructions:
1. **Input**: Accept a percentage from the user.
2. **Process**: Compare the input percentage with the grade criteria.
3. **Output**: Display the corresponding grade.


````python
marks = float(input("Enter the marks in percentage : "))

if marks >= 90:
    grade = "A+"

elif 80<= marks < 90:
    grade = "A only"
    
elif 70 <= marks < 80:
    grade = "A-"
    
elif 60 <= marks < 70:
    grade = "B+"

elif 50 <= marks < 60:
    grade = "B only"
    
elif 40 <= marks < 50:
    grade = "B-"

elif 30 <= marks <= 40:
    grade  = "C+"

else:
    grade = "Failed!"
    
print(grade)
````

### Q3. Write a program to perform arithmetic calculations on two numbers according to the user's input.

| Operator | Operation      |
|----------|----------------|
| +        | Addition       |
| -        | Subtraction    |
| *        | Multiplication |
| /        | Division       |
| **       | Exponent       |

### Instructions:
1. **Input**: Accept two numbers and an operator from the user.
2. **Process**: Perform the arithmetic operation based on the user's input.
3. **Output**: Display the result of the calculation.


````python
num1 = 16
num2 = 4

op = input("What operation you want to perform : " )

if op == "+":
    result = num1 + num2
    
elif op == "-":
    result = num1 - num2

elif op == "*":
    result = num1  * num2
    
elif op == "/":
    result = num1/num2
    
elif op == "**":
    result = num1 ** num2
    
print(result)
````

### Q4. Write a program to calculate the electricity bill by accepting the number of units from the user according to the following criteria:

| Unit        | Price        |
|-------------|--------------|
| First 100 units | No charge  |
| Next 100 units  | Rs 5 per unit |
| After 200 units | Rs 10 per unit |

### Instructions:
1. **Input**: Accept the number of units from the user.
2. **Process**: Calculate the bill amount based on the given price criteria.
3. **Output**: Display the total bill amount.

````python
bill = 99

if bill < 100:
    bill_amount = 0

elif 100 <= bill < 200:
    bill_amount = (bill - 100) * 5
    
else: 
    bill_amount = (bill - 200) * 10 + 100 * 5
    
print("Total Bill : ", bill_amount)
    
````

## Break Statement
- A break statement can be used to conditionally jump out of the loop.
- It terminates the execution of the loop.The break can be used in the while loop and for  loop. A break is mostly required, when because of some external condition, we need to exit from a loop.

```python

for i in 'machine learning':
    i == 'r':
        break
    print(i,end = ' ')
    
# Output :   m a c h i n e   l e a i n g

```

## Continue Statement 
Continue statement is used to tell Python to skip the rest of the statements of the current loop block and to move to the next iteration, of the loop.

```python
for i in 'machine learning':
    if  i == 'r':
        continue
    print(i,end = ' ')

# Output :   m a c h i n e   l e a n i n g

```

## Pass Statement
As we know, empty code is not allowed in loops, function definitions, class definitions, or if statements. Nothing happens when the pass statement is executed, but it avoids getting an error because an empty code is not allowed.

```python
	pass
```
